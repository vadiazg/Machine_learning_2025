# 1. Carga de datos:

Se carga el conjunto de datos Iris, que contiene mediciones de flores y sus respectivas etiquetas (especies).

# 2. División de datos:

 Los datos se dividen en un conjunto de entrenamiento (70%) y un conjunto de prueba (30%) para evaluar los modelos de manera objetiva.

# 3. Entrenamiento y evaluación:

*Árbol de Decisión: Se entrena un modelo de Árbol de Decisión, se hacen predicciones y se evalúa el rendimiento utilizando la precisión y un informe de clasificación detallado.

*Random Forest: Se entrena un modelo de Random Forest (con 100 árboles), se hacen predicciones y se evalúa también de la misma forma.

# 4. Visualización:

 El árbol de decisión es visualizado gráficamente para entender cómo el modelo toma decisiones basadas en las características de las flores.

El objetivo es comparar el rendimiento y la visualización de ambos modelos para ver cuál es más efectivo en la clasificación de las flores según sus características.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix

# Generar datos sintéticos para clasificación multiclase (3 clases)
X, y = make_classification(n_samples=800,
                           n_features=4,
                           n_informative=3,
                           n_redundant=0,
                           n_classes=3,
                           n_clusters_per_class=1,
                           random_state=0)

# Dividir en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=1)

# Crear un clasificador de árbol de decisión con profundidad limitada
arbol = DecisionTreeClassifier(max_depth=4, random_state=1)
arbol.fit(X_train, y_train)

# Predicciones sobre el conjunto de prueba
y_pred = arbol.predict(X_val)

# Evaluación
print("Reporte de Clasificación:\n")
print(classification_report(y_val, y_pred))

# Matriz de confusión
print("Matriz de Confusión:")
print(confusion_matrix(y_val, y_pred))

# Visualización del árbol
plt.figure(figsize=(12, 7))
plot_tree(arbol,
          filled=True,
          feature_names=['Atributo 1', 'Atributo 2', 'Atributo 3', 'Atributo 4'],
          class_names=['Clase A', 'Clase B', 'Clase C'],
          rounded=True)
plt.title("Árbol de Decisión para Clasificación Multiclase")
plt.show()

```

# Árbol de Decisión:

El Árbol de Decisión es un modelo fácil de interpretar, ya que produce una estructura visual que muestra claramente cómo toma decisiones en función de las características de los datos.

Puede ser sensible al sobreajuste (overfitting) si no se ajustan correctamente sus parámetros, aunque en este caso, el rendimiento sobre el conjunto de prueba parece ser bueno.

# Random Forest:

El Random Forest generalmente ofrece un mejor rendimiento que el Árbol de Decisión, ya que es un modelo basado en un conjunto de árboles de decisión. La combinación de varios árboles reduce la probabilidad de sobreajuste y mejora la generalización del modelo.

Random Forest es más robusto en comparación con un único Árbol de Decisión, especialmente cuando se trata de conjuntos de datos complejos o ruidosos.

# Comparación de Modelos:

En cuanto a precisión y desempeño, el Random Forest suele superar al Árbol de Decisión, ya que combina múltiples árboles para mejorar la precisión.

Ambos modelos tienen la capacidad de ofrecer interpretaciones, pero el Árbol de Decisión es más directo y visualmente comprensible, lo que es útil para presentar resultados de manera simple.

# Visualización del Árbol de Decisión:

La visualización del Árbol de Decisión es una excelente manera de entender cómo el modelo hace sus predicciones, pero puede volverse compleja si el árbol es muy grande, lo cual puede suceder si no se limitan los parámetros del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# --------------------------
# PARTE 1: Árbol de decisión en datos sintéticos multiclase
# --------------------------

# Generar datos multiclase (3 clases)
X_sint, y_sint = make_classification(n_samples=800,
                                     n_features=4,
                                     n_informative=3,
                                     n_redundant=0,
                                     n_classes=3,
                                     n_clusters_per_class=1,
                                     random_state=0)

X_train_s, X_val_s, y_train_s, y_val_s = train_test_split(X_sint, y_sint, test_size=0.3, random_state=1)

# Modelo: Árbol de Decisión
arbol = DecisionTreeClassifier(max_depth=4, random_state=1)
arbol.fit(X_train_s, y_train_s)
y_pred_s = arbol.predict(X_val_s)

# Evaluación
print("🔍 Árbol de Decisión con datos sintéticos:")
print(classification_report(y_val_s, y_pred_s))
print("Matriz de Confusión:\n", confusion_matrix(y_val_s, y_pred_s))

# Visualización del árbol
plt.figure(figsize=(12, 7))
plot_tree(arbol,
          filled=True,
          feature_names=['Atributo 1', 'Atributo 2', 'Atributo 3', 'Atributo 4'],
          class_names=['Clase A', 'Clase B', 'Clase C'],
          rounded=True)
plt.title("Árbol de Decisión - Clasificación Multiclase")
plt.show()

# --------------------------
# PARTE 2: Random Forest con el dataset Iris
# --------------------------

# Cargar Iris
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
class_names = iris.target_names

X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

# Modelo: Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train_i, y_train_i)
y_pred_i = modelo_rf.predict(X_test_i)

# Evaluación
precision_rf = accuracy_score(y_test_i, y_pred_i)
print(f"\n🌳 Random Forest con el dataset Iris:")
print(f"Precisión del modelo: {precision_rf:.2f}")
print("Reporte de Clasificación:\n", classification_report(y_test_i, y_pred_i, target_names=class_names))
print("Matriz de Confusión:\n", confusion_matrix(y_test_i, y_pred_i))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay

# --------------------------
# PARTE 1: Árbol de decisión en datos sintéticos multiclase
# --------------------------

# Generar datos multiclase (3 clases)
X_sint, y_sint = make_classification(n_samples=800,
                                     n_features=4,
                                     n_informative=3,
                                     n_redundant=0,
                                     n_classes=3,
                                     n_clusters_per_class=1,
                                     random_state=0)

X_train_s, X_val_s, y_train_s, y_val_s = train_test_split(X_sint, y_sint, test_size=0.3, random_state=1)

# Modelo: Árbol de Decisión
arbol = DecisionTreeClassifier(max_depth=4, random_state=1)
arbol.fit(X_train_s, y_train_s)
y_pred_s = arbol.predict(X_val_s)

# Evaluación
print("🔍 Árbol de Decisión con datos sintéticos:")
print(classification_report(y_val_s, y_pred_s))
print("Matriz de Confusión:\n", confusion_matrix(y_val_s, y_pred_s))

# Matriz de confusión para Árbol de Decisión
cm_arbol = confusion_matrix(y_val_s, y_pred_s)
disp_arbol = ConfusionMatrixDisplay(confusion_matrix=cm_arbol, display_labels=['Clase A', 'Clase B', 'Clase C'])
disp_arbol.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión - Árbol de Decisión (Datos Sintéticos)")
plt.show()

# Visualización del árbol
plt.figure(figsize=(12, 7))
plot_tree(arbol,
          filled=True,
          feature_names=['Atributo 1', 'Atributo 2', 'Atributo 3', 'Atributo 4'],
          class_names=['Clase A', 'Clase B', 'Clase C'],
          rounded=True)
plt.title("Árbol de Decisión - Clasificación Multiclase")
plt.show()

# --------------------------
# PARTE 2: Random Forest con el dataset Iris
# --------------------------

# Cargar Iris
iris = load_iris()
X_iris = iris.data
y_iris = iris.target
class_names = iris.target_names

X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)

# Modelo: Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train_i, y_train_i)
y_pred_i = modelo_rf.predict(X_test_i)

# Evaluación
precision_rf = accuracy_score(y_test_i, y_pred_i)
print(f"\n🌳 Random Forest con el dataset Iris:")
print(f"Precisión del modelo: {precision_rf:.2f}")
print("Reporte de Clasificación:\n", classification_report(y_test_i, y_pred_i, target_names=class_names))
print("Matriz de Confusión:\n", confusion_matrix(y_test_i, y_pred_i))

# Matriz de confusión para Random Forest
cm_rf = confusion_matrix(y_test_i, y_pred_i)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=iris.target_names)
disp_rf.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión - Random Forest (Iris)")
plt.show()
```

# El siguiene  código se centra en la curva ROC y el AUC (Área bajo la Curva), que son herramientas muy útiles para evaluar el rendimiento de los modelos de clasificación, especialmente cuando se tienen múltiples clases (como en el caso del conjunto de datos Iris, que tiene tres clases).

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Convertimos las clases a formato binarizado (para multiclasificación)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])  # Convertimos las clases en 0, 1, 2
y_pred_dt_bin = label_binarize(y_pred_dt, classes=[0, 1, 2])
y_pred_rf_bin = label_binarize(y_pred_rf, classes=[0, 1, 2])

# Curvas ROC para el Árbol de Decisión
fpr_dt, tpr_dt, _ = roc_curve(y_test_bin.ravel(), y_pred_dt_bin.ravel())
roc_auc_dt = auc(fpr_dt, tpr_dt)

# Curvas ROC para Random Forest
fpr_rf, tpr_rf, _ = roc_curve(y_test_bin.ravel(), y_pred_rf_bin.ravel())
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Graficamos las curvas ROC
plt.figure(figsize=(10,6))
plt.plot(fpr_dt, tpr_dt, color='blue', lw=2, label=f'Árbol de Decisión (AUC = {roc_auc_dt:.2f})')
plt.plot(fpr_rf, tpr_rf, color='green', lw=2, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC - Árbol de Decisión vs Random Forest')
plt.legend(loc='lower right')
plt.show()
```

En el grafico anterior se trazan las curvas ROC de ambos modelos (Árbol de Decisión y Random Forest) en la misma gráfica. Las curvas ROC muestran la relación entre la tasa de falsos positivos y la tasa de verdaderos positivos para cada modelo a diferentes umbrales de clasificación. El área bajo cada curva (AUC) también se muestra en la leyenda.

 Proporción de ejemplos negativos que son clasificados incorrectamente como positivos.

# Tasa de Verdaderos Positivos (TPR):
